Imports

```python
import pandas as pd
import numpy as np
```

Definitions

```python
file_meta = "metadata_files.csv"
file_kfolds = "split_kfolds.csv"
file_meta_singers = "metadata_singers.csv"

#join the singers metadata and the file metadata dataframes: returns 0 if there is no corresponding technique and range in the singers metadata
def get_singer_rank(row):
    col_name = row['meta_singers_col']
    if col_name in row.index:
        return row[col_name]
    else:
        return 0
```

Open metadata

```python
df_meta = pd.read_csv(file_meta)
df_k_folds = pd.read_csv(file_kfolds)
df_meta_singers = pd.read_csv(file_meta_singers)
```

Merge k folds metadata with file metadata

```python
df_meta = df_meta.merge(df_k_folds, on='file_name')
```

Merge medata of files with metadata of singers

```python
#merge the metadata of files with the metadata of singers
df_meta = df_meta.merge(df_meta_singers, on='singer_id')

#create a new column in df_meta that matches the techniques/range in df_meta_singers
col_singers_techniques = df_meta_singers.columns[-14:].tolist()
joined_name_range = []

for item in col_singers_techniques:
    parts = item.split('_')
    if len(parts) == 1:
        new_item = f"{item}_-"
    else:
        new_item = item
    joined_name_range.append(new_item)

association_dict = {}

for new_item, original_item in zip(joined_name_range, col_singers_techniques):
    association_dict[new_item] = original_item

df_meta['meta_singers_col'] = (df_meta['name'] + '_' + df_meta['range']).map(association_dict)

#create a singer rank column 
df_meta['singer_rank'] = df_meta.apply(get_singer_rank, axis=1)

#drop the useless columns
df_meta.drop(columns=col_singers_techniques, inplace=True)
df_meta.drop(columns='meta_singers_col', inplace=True)

df_meta
```
